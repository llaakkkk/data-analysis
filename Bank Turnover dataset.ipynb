```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_women = bank_df[bank_df['Gender'] == 'Female']

# Print the length of the new DataFrame
print(len(df_women))

# Initialize the figure and add subplots
fig = plt.figure(figsize=(15, 12))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# Create distribution plots for Hits, Home Runs, Years Played and All Star Games
ax1.hist(df_women['Age'])
ax1.set_title('Distribution of Age')
ax1.set_ylabel('Women Age')
ax2.hist(df_women['Tenure'])
ax2.set_title('Distribution of Tenure')
ax3.hist(df_women['NumOfProducts'])
ax3.set_title('Distribution of Numbers Of Products')
ax3.set_ylabel('Women Numbers Of Products')
ax4.hist(df_women['CreditScore'])
ax4.set_title('Distribution of Credit Score')

# Show the plot
plt.show()
```

```python
df_men = bank_df[bank_df['Gender'] == 'Male']

# Print the length of the new DataFrame
print(len(df_men))

# Initialize the figure and add subplots
fig = plt.figure(figsize=(15, 12))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# Create distribution plots for Hits, Home Runs, Years Played and All Star Games
ax1.hist(df_men['Age'])
ax1.set_title('Distribution of Age')
ax1.set_ylabel('Men Age')
ax2.hist(df_men['Tenure'])
ax2.set_title('Distribution of Tenure')
ax3.hist(df_men['NumOfProducts'])
ax3.set_title('Distribution of Numbers Of Products')
ax3.set_ylabel('Men Numbers Of Products')
ax4.hist(df_men['CreditScore'])
ax4.set_title('Distribution of Credit Score')

# Show the plot
plt.show()
```

```python
gender_df = pd.get_dummies(bank_df['Gender'], prefix='gender')
bank_df = pd.concat([bank_df, gender_df], axis=1)
```
